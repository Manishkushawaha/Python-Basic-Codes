```python
cdf_one_minus_9 = 1-stats.uniform(loc = 3,scale = 10).cdf(9)  
```

```python
cdf_one_minus_9
```

```python
cdf_7 = stats.uniform(loc = 3,scale = 10).cdf(7)
```

```python
seven_to_nine = (1-(cdf_7+cdf_one_minus_9))
seven_to_nine
```

```python
cdf_2 = stats.uniform(loc = 1,scale = 10).cdf(2)
cdf_2
```

```python
cdf_one_minus_6 = 1-stats.uniform(loc = 1,scale = 10).cdf(6)  
cdf_one_minus_6
```

```python
two_to_six = (1-(cdf_2+cdf_one_minus_6))
two_to_six
```
