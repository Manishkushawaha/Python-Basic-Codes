Pandas GroupBy Function in Python

Pandas GroupBy function is used to split the data into groups based on some criteria.

Any GroupBy operation involves one of the following operations on the original object:

-Splitting the object

-Applying a function

-Combining the result

Syntax: DataFrame.groupby()

```python
df=pd.read_csv('D:\\Data sets\\student_result2.csv')
df
```

```python
gr1=df.groupby(by='Section')
gr1
```

```python
gr1.groups
```

```python
df.groupby(['Section','Class']).groups
```

```python
for Class in df:
    for Class in gr1:
        print(df)
        print(Class)
```

# Join method

DataFrame join is a convenienyt method for combing the columns of two potentially differently_indexed

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
```

```python
df2 = pd.DataFrame({'C': [4,5,6],
                   'D': [40,50,60]})
 
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]},
                   index = ['a','b','c'])
 
df2 = pd.DataFrame({'C': [4,5,6],
                   'D': [40,50,60]},
                  index = ['a','b','c'])
 
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]},
                   index = ['a','b','c'])
 
df2 = pd.DataFrame({'C': [4,5],
                   'D': [40,50]},
                  index = ['a','b'])
 
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
df2 = pd.DataFrame({'A': [4,5,6],
                   'D': [40,50,60]})
 
display(df1, df2)
```

# Append function

Pandas append function is used to append rows of other dataframe to the end of the given dataframe,returning a new dataframe object

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
 
df2 = pd.DataFrame({'A': [4,5,6],
                   'B': [40,50,60]})
 
display(df1 ,df2)
```

```python
pd.concat([df1,df2])
```

```python
df1.append(df2,ignore_index=True)
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
 
df2 = pd.DataFrame({'C': [4,5,6],
                   'B': [40,50,60]})
 
display(df1 ,df2)
```

```python
df1.append(df2, ignore_index = True)
```

```python
df1.append(df2, ignore_index = True, sort = False)
```

# Pivot Table

The levels in the  pivot table will be sorted in multi index object(hierarchicalindexes) on the index and columns of the result df

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Data sets\\Movie_rating.csv')
df
```

```python
df.pivot_table(index='Genre')
```

```python
df.pivot_table(index='Genre',columns='Year')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='count')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='max')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='max',fill_value='none')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='max',fill_value=0)
```
