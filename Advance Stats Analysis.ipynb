```python
from scipy.stats import  binom
import matplotlib.pyplot as plt
import seaborn as sns


# Define the parameters for the binomial distribution
n = 100
p = 0.05

# Create an array of possible outcomes
x = range(n+1)

# Calculate the probability mass function for each outcome
binom = binom.rvs( n, p,size=1000)

# Plot the probability mass function
plt.figure(figsize=(10,5))
sns .set_style("darkgrid")

sns.histplot(binom,bins=x,kde=True,color="green")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial Probability Mass Function")
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

heights = [165, 170, 172, 175, 168, 173, 167, 170, 169, 171, 172, 174, 170, 168, 172, 167, 170, 171, 172, 175, 173, 170, 170, 168, 169]

# Set the population mean
pop_mean = 170

# Compute the t-statistic and p-value
t_statistic, p_value = stats.ttest_1samp(heights, pop_mean)

# Set the significance level
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")


```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Set the population mean IQ score
pop_mean = 100

# Set the sample size and the sample mean IQ score
n = 30
sample_mean = 105

# Set the sample standard deviation
sample_std = 15

# Compute the t-statistic and p-value
t_statistic, p_value = stats.ttest_1samp((sample_mean - pop_mean) / (sample_std / np.sqrt(n)), 0)

# Set the significance level
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

# Plot the distribution of the t-statistic with the critical values and the t-statistic for the sample
t_values = np.linspace(-4, 4, 1000)
plt.plot(t_values, stats.t.pdf(t_values, n-1), 'k-', label='t-distribution')
plt.axvline(stats.t.ppf(alpha/2, n-1), color='g', linestyle='--', label='Lower critical value')
plt.axvline(stats.t.ppf(1-alpha/2, n-1), color='g', linestyle='--', label='Upper critical value')
plt.axvline(t_statistic, color='r', linestyle='--', label='Sample t-statistic')
plt.legend()
plt.xlabel('t')
plt.ylabel('Probability density')
plt.show()
```
